```python
import pandas as pd
from transformers import AutoTokenizer, pipeline

# Load tokenizer and model once
tokenizer = AutoTokenizer.from_pretrained("google/flan-t5-small")
feature_extractor = pipeline("text2text-generation", model="google/flan-t5-small")

def extract_features_local(text, max_length=512):
    inputs = tokenizer(text, truncation=True, max_length=512, return_tensors="pt")
    truncated_text = tokenizer.decode(inputs["input_ids"][0], skip_special_tokens=True)
    prompt = f"Extract product features mentioned in the review:\n\"{truncated_text}\"\nFeatures:"
    outputs = feature_extractor(prompt, max_new_tokens=50)
    return outputs[0]['generated_text']

# Load datasets and reviews
df5 = pd.read_csv('/content/archive_5/Reviews.csv')
reviews_5 = df5['Text'].dropna().tolist()

df6 = pd.read_csv('/content/archive_6/Reviews.csv')
reviews_6 = df6['Text'].dropna().tolist()

df7 = pd.read_csv('/content/archive_7/Dataset-SA.csv')
reviews_7 = df7['Review'].dropna().tolist()

df8_1 = pd.read_csv('/content/archive_8/all_kindle_review.csv')
reviews_8_1 = df8_1['reviewText'].dropna().tolist()

df8_2 = pd.read_csv('/content/archive_8/preprocessed_kindle_review.csv')
reviews_8_2 = df8_2['reviewText'].dropna().tolist()

def process_and_save(reviews, filename, count=100):
    results = []
    for review in reviews[:count]:
        features = extract_features_local(review)
        results.append({'Review': review, 'Features': features})
    df_features = pd.DataFrame(results)
    df_features.to_csv(filename, index=False)
    print(f"Saved features to {filename}")

process_and_save(reviews_5, '/content/extracted_features_5.csv')
process_and_save(reviews_6, '/content/extracted_features_6.csv')
process_and_save(reviews_7, '/content/extracted_features_7.csv')
process_and_save(reviews_8_1, '/content/extracted_features_8_1.csv')
process_and_save(reviews_8_2, '/content/extracted_features_8_2.csv')
```
